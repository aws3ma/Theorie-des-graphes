```python
#Ce algorithme permet de retourner les sommets connecter
def Parcour_en_Profendeur_sommet_connecter(graph,début):
    visité=[]
    pile =[début]
    while pile:
        sommet = pile.pop()
        visité.append(sommet)
        pile +=graph[sommet]
    return visité
```

```python
#Ce algorithme permet de retourner le chemin entre deux sommets s'il existe
def Parcour_en_Profendeur_chemin(graph,début,fin,chemin=None):
    if chemin==None:
        chemin = list(début)
    if(début==fin):
        yield chemin
    else:
        childs=graph[début]
        for child in childs:
            yield from Parcour_en_Profendeur_chemin(graph,child,fin,chemin+ [child])
            
```
